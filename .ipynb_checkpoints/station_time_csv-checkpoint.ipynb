```python
# API 관련 변수
time_secret_key = '46686b666579756e38314c6a554a4c'
time_urls = "http://openapi.seoul.go.kr:8088"

# 전철역 코드 데이터 불러오기
train_code = pd.read_csv('station_code.csv', encoding='EUC-KR')

# 02호선 데이터만 필터링
line_02 = train_code[train_code['호선'] == '02호선']

# 방향과 요일 설정
directions = [1, 2]  # 내선 - 1, 외선 - 2
days = [1, 2, 3]  # 평일-1, 토요일-2, 일요일/공휴일-3

for day in days:
    for direction in directions:
        time_tables = []
        for index, row in line_02.iterrows():
            train_station = row['전철역코드']
            time_url = f'{time_urls}/{time_secret_key}/xml/SearchSTNTimeTableByIDService/1/1000/{train_station}/{day}/{direction}'
            response_time_table = requests.get(time_url)
            soup_time_table = BeautifulSoup(response_time_table.text, "xml")
            arrival_times = [tag.text for tag in soup_time_table.find_all("ARRIVETIME")]

            # 역별 시간표 저장
            time_tables.append({
                '역코드': row['전철역코드'],
                '역명': row['전철역명'],
                '시간표': ', '.join(arrival_times)
            })

        # 결과를 DataFrame으로 변환
        time_tables_df = pd.DataFrame(time_tables)

        # CSV 파일로 저장 (요일과 방향에 따라 파일명 지정)
        day_str = "평일" if day == 1 else "토요일" if day == 2 else "일요일_공휴일"
        direction_str = "내선" if direction == 1 else "외선"
        filename = f'line_02_time_tables_{day_str}_{direction_str}.csv'
        time_tables_df.to_csv(filename, index=False)

        print(f"파일 저장 완료: {filename}")
```
