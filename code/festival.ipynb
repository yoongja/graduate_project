```python
import pandas as pd
```

```python
festival_data = pd.read_csv('seoul_festival.csv',encoding='cp949', encoding_errors='ignore')
festival_data
```

```python
festival_data.head()
```

```python
selected_festival = festival_data[['공연/행사명','분류','시작일', '종료일', '장소', '위도(X좌표)', '경도(Y좌표)']]

# 선택된 열을 출력
selected_festival
```

```python
station = pd.read_excel('all_train.xlsx')
station.keys()
```

```python
select_station = station[['역번호','역사명','노선번호','역위도','역경도']]
select_station
```

```python
import math
```

```python
# 하버사인 함수 정의
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    km = 6371 * c
    return km * 1000

# 행사 장소 데이터
festival_data = selected_festival

# 역 정보 데이터
station_data = select_station

results = []

# 각 행사 장소마다 가장 가까운 역 찾기
for index, festival in festival_data.iterrows():
    closest_station = None
    min_distance = float('inf')
    for _, station in station_data.iterrows():
        dist = haversine(festival['위도(X좌표)'], festival['경도(Y좌표)'], station['역위도'], station['역경도'])
        if dist < min_distance:
            min_distance = dist
            closest_station = station
            
    # 가장 가까운 역이 없는 경우 처리
    if closest_station is None:
        closest_station_name = '역 없음'
        min_distance = -1
    else:
        closest_station_name = closest_station['역사명']

    results.append({
        '분류' : festival['분류'],
        '행사명' : festival['공연/행사명'],
        '시작일': festival['시작일'],
        '종료일': festival['종료일'],
        '장소': festival['장소'],
        '가장 가까운 역': closest_station_name,
        '거리(미터)': min_distance
    })

# 결과를 DataFrame으로 변환
results_df = pd.DataFrame(results)

# CSV 파일로 저장
results_df.to_csv('nearest_stations.csv', index=False)
```
